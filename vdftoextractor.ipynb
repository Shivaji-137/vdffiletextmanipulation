```python
import re

# Define the path to the file
file_path = '/home/shivaji/Documents/vdffiletextmanipulation/output/freqgas.vdf'

extracted_pairs = []
pattern2 = r"^s\s\s?\d+\s+\w+\s+\w+"#r"^s\s\d{1,3}\s+\w+\s+\w+"
another_pairs = []
# Read the file and extract the required pairs
with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
    for line in file:
        match2 = re.findall(pattern2, line)
        matches = re.findall(r'(s\d+)\s+(-?\d+)', line)
       
        if matches != []:
            # print(type(matches))
            extracted_pairs.append(matches)
        if match2 != []:
            another_pairs.append(match2[0])
            
            

extracted_pairs
```

```python
reextracted = []
for i in extracted_pairs:
    res = " "
    for j in i:
        a,b = j
        init = a + f"({abs(int(b))})"
        if abs(int(b)) > 10:
            res = res + " "+ init
        
    reextracted.append(res.lstrip())
reextracted
```

```python
dicval = {}
for j in another_pairs:
    lsplit = j.split("   ",1)
    key = "".join(lsplit[0].split())
    val = lsplit[1]
    dicval[key] = val
dicval
```

```python
def modify(reextracted, dicvals):
    keys = list(dicvals.keys())
    xval = []
    for j in reextracted:
        p = j
        for k in keys:
            if k in j:
                z = re.sub(rf"\b{k}\b", dicvals.get(k), j)
                p = z
        xval.append(p)
    return xval
    
```

```python
onemod = modify(reextracted, dicval)
onemod
```

```python
secondmod = modify(onemod, dicval)
thirdmod = modify(secondmod, dicval)
fourthmod = modify(thirdmod, dicval)
fourthmod
```

```python
fifthmod = modify(fourthmod, dicval)
fifthmod
```
